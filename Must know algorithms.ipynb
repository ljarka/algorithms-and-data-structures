## Binary search

```kotlin
val input = listOf(0,1,3,4,5,6,7,8,9,10, 55)


fun binarySearch(input: List<Int>, valueToFind: Int): Int {
    var l = 0
    var r = input.size - 1 
    
    while (l <= r) {
        var middle = l + (r - l) / 2
        
        if (valueToFind > input[middle]) {
            l = middle + 1
        } else if (valueToFind < input[middle]) {
            r = middle - 1
        } else {
            return middle
        }
        
    }
    
    return  -1
}

binarySearch(input, 55)
```

## BFS

```kotlin
class TreeNode<T>(val value: T) {
    var left: TreeNode<T>? = null
    var right: TreeNode<T>? = null
}

val root = TreeNode(0)
val left = TreeNode(1)
val right = TreeNode(2)

root.left = left
root.right = right


fun bfs(root: TreeNode<Int>) {
    val queue = mutableListOf<TreeNode<Int>>()
    queue.add(root)
    
    while(queue.isNotEmpty()) {
        val node = queue.removeFirst()
        println(node.value)
        
        if (node.left != null) {
            queue.add(node.left!!)
        }
        
        if (node.right != null) {
            queue.add(node.right!!)
        }
    }
}

bfs(root)
```

## BFS Graphs 

```kotlin
class Node<T>(val value: T) {
    var children : List<Node<T>>? = null
}

fun bfsGraph(root: Node<Int>) {
    val visited = mutableSetOf<Node<Int>>()
    val queue = mutableListOf<Node<Int>>()
    queue.add(root)
    visited.add(root)
    
    while (queue.isNotEmpty()) {
        val currentNode = queue.removeFirst()
        
        println(currentNode.value)
        
        val children = currentNode.children
        
        if (children != null) {
             for(node in children){
                 if (node !in visited) {
                     visited.add(node)
                     queue.add(node)
                 }
             }
        }
        
    }
}

val graph = Node(0)
graph.children = listOf(Node(1), Node(2), Node(3))

bfsGraph(graph)
```

## DFS

```kotlin
fun dfs(root: TreeNode<Int>) {
    val stack = mutableListOf<TreeNode<Int>>()
    stack.add(root)
    
    while(stack.isNotEmpty()) {
        val node = stack.removeLast()
        println(node.value)
              
        if (node.right != null) {
            stack.add(node.right!!)
        }
        
        if (node.left != null) {
            stack.add(node.left!!)
        }
    }
}

dfs(root)
```

```kotlin
fun dfsRecursive(node: TreeNode<Int>?) {
    if (node == null) return  
    
    println(node.value)
    
    dfsRecursive(node.left)
    dfsRecursive(node.right)
}

dfsRecursive(root)
```

## DFS graph

```kotlin
class Node<T>(val value: T, val children : List<Node<T>>? = null)

fun dfsGraph(root: Node<Int>) {
    val stack = mutableListOf<Node<Int>>()
    val visited = mutableSetOf<Node<Int>>()
    stack.add(root)
    visited.add(root)
    
    while(stack.isNotEmpty()) {
        val currentElement = stack.removeLast()
        println(currentElement)
        val children = currentElement.children
        
        if (children != null) {
            for(child in children.reversed()) {
                if (child !in visited) {
                    visited.add(child)
                    stack.add(child)
                }
            }
        }
    }
}


val graph = Node(0, listOf(Node(1), Node(2), Node(3)))

dfsGraph(graph)
```

## DFS graph recursive

```kotlin
class Node<T>(val value: T, val children : List<Node<T>>? = null)

fun dfsGraphRecursive(node: Node<Int>? = null, visited: MutableList<Node<Int>>) {
    if (node == null) {
        return
    }
   
    println(node.value)
    val children = node.children
        
    if (children != null) {
        for(child in children) {
            if (child !in visited) {
                visited.add(child)
                dfsGraphRecursive(child, visited)
            }
        }
    }
}


val graph = Node(0, listOf(Node(1), Node(2), Node(3)))

dfsGraphRecursive(graph, mutableListOf(graph))
```

## HEAP

```kotlin
import java.util.PriorityQueue

val minHeap = PriorityQueue<Int>(compareByDescending { it })
minHeap.addAll(listOf(5, 1, 8, 3, 2))
while (minHeap.isNotEmpty()) {
  print("${minHeap.poll()} ")   // 1 2 3 5 8
}
```
