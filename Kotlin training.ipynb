## Stack implementation 

```kotlin
class Stack<T> {
    private val items = mutableListOf<T>()
    
    fun push(item: T) {
        items.add(item)
    }
    
    fun pop(): T? {
        if (items.isEmpty()) return null 
        
        return items.removeLast()
    }
    
    fun peek(): T? {
        if (items.isEmpty()) return null 
        
        return items.last()
    }
}
```

```kotlin
val stack = Stack<Int>()
stack.push(1)
stack.push(2)
stack.push(3)

println(stack.pop())
println(stack.pop())
```

## Basic LinkedList

```kotlin
class ListItem<T>(val value: T) {
    var next: ListItem<T>? = null
}

class LinkedList<T> {
    var head: ListItem<T>? = null
    
    fun addFirst(value: T) {
        val newHead = ListItem(value)
        newHead.next = head
        head = newHead
    }

    fun getFirst() = head?.value
}

val linkedList = LinkedList<Int>() 
linkedList.addFirst(1)
linkedList.addFirst(2)
linkedList.addFirst(3)

var current = linkedList.head

while (current != null) {
    println(current?.value)
    current = current?.next
}
```

## LinkedList implementation 

```kotlin
class ListItem<T>(val value: T) {
    var next: ListItem<T>? = null
}


class LinkedList<T> {
    private var size: Int = 0
    private var head: ListItem<T>? = null
    private var tail: ListItem<T>? = null
    
    fun add(value: T) {
        if (tail == null) {
            val item = ListItem(value)
            head = item
            tail = item
        } else {
            tail?.next = ListItem(value)
            tail = tail?.next
        }
        
        size++
    }
    
    fun getLast() = tail?.value
    
    fun getFirst() = head?.value
    
    fun removeFirst() {
        if (head == null) return
        
        head = head?.next
        
        if (head == null) {
            tail = null        
        }
        size--
    }
    
    fun getSize() = size
}
```

## Iterate over matrix

```kotlin
var matrix = Array(9){IntArray(9)}

var x = 0
var counter = 0 
while (x < matrix.size){
    var y = 0
    while (y < matrix.get(0).size){
        matrix[x][y] = counter
        counter++
        y++
    }
    
    x++
}

matrix.forEach { row -> 
    val formattedRow = row.map { 
       if (it < 10) {
         "0$it"
       }else {
         it.toString()
       }
   }
   println(formattedRow.toString())
}
```

```kotlin
fun isTrionic(nums: IntArray): Boolean {
        val increasingArray = IntArray(3) // -1 when decreasing, +1 when increasing
        // increasing, decreasing, increasing 

        var currentIndexOfResult = -1
        var currentDirection = 0 
        var prevValue : Int? = null // 6
        
        for (num in nums){
            if (currentIndexOfResult > 2) {
                println("currentIndexOfResult " + currentIndexOfResult)
                return false
            }

            if (prevValue == null) {
                prevValue = num
                continue
            } else if (num > prevValue) {
                if (currentDirection != 1) {
                    currentIndexOfResult++
                    currentDirection = 1
                    increasingArray[currentIndexOfResult] = 1   
                }
            } else if (num < prevValue) {
                if (currentDirection != -1) {
                    currentIndexOfResult++
                    currentDirection = -1
                    increasingArray[currentIndexOfResult] = -1   
                }
            }    
            prevValue = num
        }

        println(increasingArray.joinToString())
        
        return increasingArray[0] == 1 && increasingArray[1] == -1 && increasingArray[2] == 1
}


isTrionic(intArrayOf(6,8,2,5,9))
```
