## Classic Binary search

## Is `target` inside this sorted array?

```kotlin
fun binarySearch(array: IntArray, target: Int): Int {
    var l = 0 
    var r = array.lastIndex
    
    while (l <= r) {
        val mid = l + (r - l) / 2
        
        when {
            array[mid] > target -> r = mid - 1
            array[mid] < target -> l = mid + 1
            else -> return mid
        }
    }
    return -1
}
```

```kotlin
val array = intArrayOf(1, 3, 5, 7, 9)
println(binarySearch(array, 5)) // 2
println(binarySearch(array, 6)) // - 1
```

## Lower bound binary search

### Where is the first index whose value is greater than or equal to `target`?

```kotlin
fun lowerBound(array: IntArray, target: Int) : Int {
    var l = 0
    var r = array.size // r bound is exclusive
    
    while (l < r) { 
        val mid = l + (r - l) / 2
        
        if (array[mid] >= target) {
            r = mid
        } else {
            l = mid + 1
        }
    }
    
    return l // return array size when not found 
}
```

```kotlin
val array = intArrayOf(1, 3, 3, 3, 5, 7)

println(lowerBound(array, 3)) //1
println(lowerBound(array, 4)) //4
println(lowerBound(array, 8)) //6
```

## Binary search on answer

### I am not searching in an array. I ma searching for the smallest/largest possible answer

```kotlin
fun minEatingSpeed(piles: IntArray, h: Int) : Int {
    var l = 1
    var r = piles.max()
    
    while (l < r) {
        val mid = l + (r - l) / 2
        
        if (hours(mid, piles) <= h) {
           r = mid 
        } else {
           l = mid + 1
        }
    }   
    return l
}

private fun hours(speed: Int, piles: IntArray) : Int {
    var time = 0
    
    for (pile in piles) {
        time += ceil(pile / speed.toDouble()).to()
    }
    
    return time
}
```

```kotlin
val piles = intArrayOf(3, 6, 7, 11)
val h = 8

println(minEatingSpeed(piles, h)) // 4
```
