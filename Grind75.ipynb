## 1. Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

https://leetcode.com/problems/two-sum/

```kotlin
class Solution {
    fun twoSum(nums: IntArray, target: Int): IntArray {
        val values = mutableMapOf<Int, Int>()
        
        for (i in nums.indices) {
            val value = target - nums[i]

            if (!values.contains(value)) {
                values.put(nums[i], i)
            } else {
                return intArrayOf(i, values.get(value) ?: 0)  
            }
        }

        return intArrayOf()
    }
}
val array = intArrayOf(2,7,11,15)
val target = 9
val result = Solution().twoSum(array, target)
assert(result == intArrayOf(1,0)) { "Expected [1, 0]" }
println("Success")
```

## 2. Valid Parentheses
Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.

https://leetcode.com/problems/valid-parentheses

```kotlin
class Solution {
    val opening = setOf('(', '{', '[')
    val closing = setOf(')', '}', ']')
    val pairs = mapOf('(' to ')', '{' to '}', '[' to ']')
    
    fun isValid(s: String): Boolean {
        val stack = mutableListOf<Char>() 

        s.forEach{
            if (it.isOpening()) {
                stack.add(it)
            } else {
                val last = stack.removeLastOrNull()  

                if (pairs.get(last) != it) {
                    return false
                } 
            }
        }

        return stack.isEmpty()
    }

    private fun Char.isOpening() : Boolean = opening.contains(this)
}

val result = Solution().isValid("([])")

assert(result) { "Expected true" }
println("Success")
```

## 3. Merge Two Sorted Lists

You are given the heads of two sorted linked lists list1 and list2.
Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.
Return the head of the merged linked list.

https://leetcode.com/problems/merge-two-sorted-lists

```kotlin
data class ListNode(val value: Int?, var next: ListNode? = null)

class Solution {
    fun mergeTwoLists(list1: ListNode?, list2: ListNode?): ListNode? {
        var list1CurrentNode = list1
        var list2CurrentNode = list2

        var resultHead : ListNode? = null
        var resultCurrentNode: ListNode? = null

        while (list1CurrentNode != null || list2CurrentNode != null) {
            val list1CurrentNodeValue = list1CurrentNode?.value ?: Int.MAX_VALUE
            val list2CurrentNodeValue = list2CurrentNode?.value ?: Int.MAX_VALUE
            
            if (list1CurrentNodeValue > list2CurrentNodeValue) {
                if (resultHead == null) {
                    resultHead = list2CurrentNode
                    resultCurrentNode = resultHead
                } else {
                    resultCurrentNode?.next = list2CurrentNode
                    resultCurrentNode = resultCurrentNode?.next
                } 

                list2CurrentNode = list2CurrentNode?.next
            } else {
                if (resultHead == null) {
                    resultHead = list1CurrentNode
                    resultCurrentNode = resultHead
                } else {
                    resultCurrentNode?.next = list1CurrentNode
                    resultCurrentNode = resultCurrentNode?.next
                }
                list1CurrentNode = list1CurrentNode?.next
            }
        }
        
        return resultHead
    }
}

fun creatLinkedList(vararg args: Int) : ListNode? {
    var head : ListNode? = null
    var currentNode : ListNode? = null
    
    args.forEach {
        if (head == null) {
            head = ListNode(it)
            currentNode = head
        } else {
            currentNode?.next = ListNode(it)
            currentNode = currentNode?.next
        }
    }

    return head
}

fun printLinkedList(listNode: ListNode?) {
    var currentNode = listNode

    while (currentNode != null) {
        currentNode?.let {
            println(it.value)
        }
        currentNode = currentNode?.next
    }
}

val list = creatLinkedList(1,2,3,4)

val list2 = creatLinkedList(1,3,5,6)


printLinkedList(Solution().mergeTwoLists(list, list2))
```

## 4 Best Time to Buy and Sell Stock
You are given an array prices where prices[i] is the price of a given stock on the ith day.
You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.
Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.


```kotlin
class Solution {
    fun maxProfit(prices: IntArray): Int {
        var maxProfit = 0
        var buy = Int.MAX_VALUE
        
        for (price in prices) {
            if (price < buy) {
                buy = price 
            }
            maxProfit = maxOf(price - buy, maxProfit) 
        } 
        
        return maxProfit
    }
}

val result = Solution().maxProfit(intArrayOf(7,1,5,3,6,4))

assert(result == 5) { "Expected true" }
result
```
